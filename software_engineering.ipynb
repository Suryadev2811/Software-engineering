```python
def quadratic_solution(a, b, time, c):
    discriminant = b**2 - 4*a*(c - time)
    sol1 = (-b - discriminant**0.5) / (2*a)
    sol2 = (-b + discriminant**0.5) / (2*a)
    return sol1, sol2

def weather_model(mode):
    if mode == "hardcoded":
        a, b, c = 0.5, -2.0, 20.0
        for time in range(1, 4):
            temp = quadratic_solution(a, b, time, c)[1]
            print(f"Temperature at time {time}: {temp:.2f} degrees")
    elif mode == "input":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter initial temperature: "))
        for time in range(1, 4):
            temp = quadratic_solution(a, b, time, c)[1]
            print(f"Temperature at time {time}: {temp:.2f} degrees")
    elif mode == "file":
        try:
            with open('spam.txt', 'r') as file:
                for set_num, line in enumerate(file):
                    try:
                        a, b, c = [float(value) for value in line.strip().split() if value]
                        print(f"Weather simulation for set {set_num + 1}:")
                        for time in range(1, 4):
                            temp = quadratic_solution(a, b, time, c)[1]
                            print(f"  Temperature at time {time}: {temp:.2f} degrees")
                    except ValueError:
                        print(f"Warning: Invalid data on line {set_num + 1}. Skipping.")
        except FileNotFoundError:
            print("Error: spam.txt not found.")

# Choose the mode of operation
mode = input("Enter mode ('hardcoded', 'input', or 'file'): ").lower()
weather_model(mode)
```

```python
def calculate_weather_prediction(temperature, humidity, wind_speed):
    """Calculates the weather prediction (W) using the quadratic model."""
    W = 0.5 * (temperature**2) - 0.2 * humidity + 0.1 * wind_speed - 15
    return W

def map_prediction_to_category(W):
    """Maps the calculated W value to a weather category."""
    if W > 25:
        return "Sunny"
    elif 15 < W <= 25:
        return "Cloudy"
    elif 5 < W <= 15:
        return "Rainy"
    else:
        return "Stormy"

def get_weather_prediction(temperature, humidity, wind_speed):
    """Gets the weather prediction and category."""
    W = calculate_weather_prediction(temperature, humidity, wind_speed)
    category = map_prediction_to_category(W)
    return category, W  # Return both the category and W value


def main():
    """Main function to handle user input and output."""
    mode = input("Enter input mode ('manual' or 'file'): ").lower()

    if mode == "manual":
        temperature = float(input("Enter temperature: "))
        humidity = float(input("Enter humidity: "))
        wind_speed = float(input("Enter wind speed: "))
        category, W = get_weather_prediction(temperature, humidity, wind_speed)
        print(f"Weather prediction: {category} (W = {W:.2f})")

    elif mode == "file":
        try:
            with open("spam.txt", "r") as file:
                for line_num, line in enumerate(file, 1):
                    try:
                        data = [float(value) for value in line.strip().split()]
                        if len(data) == 3:  # Make sure there are 3 values per line
                            temperature, humidity, wind_speed = data
                            category, W = get_weather_prediction(temperature, humidity, wind_speed)
                            print(f"Weather prediction for line {line_num}: {category} (W = {W:.2f})")
                        else:
                            print(f"Warning: Invalid data on line {line_num}. Skipping.")
                    except ValueError:
                        print(f"Warning: Invalid data on line {line_num}. Skipping.")

        except FileNotFoundError:
            print("Error: spam.txt not found.")

    else:
        print("Invalid input mode. Please enter 'manual' or 'file'.")

if __name__ == "__main__":
    main()
```

```python
def quadratic_solution(a, b, time, c):
    discriminant = b**2 - 4*a*(c - time)
    if discriminant < 0:
        raise ValueError("Negative discriminant, no real solutions.")
    sol1 = (-b - discriminant**0.5) / (2*a)
    sol2 = (-b + discriminant**0.5) / (2*a)
    return sol1, sol2

def weather_model():
    print("Welcome to the Incremental Weather Model Simulation!")
    mode = input("Choose mode ('hardcoded', 'input', or 'file'): ").lower()

    if mode == "hardcoded":
        a, b, c = 0.5, -2.0, 20.0
    elif mode == "input":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter initial temperature: "))
    elif mode == "file":
        try:
            with open('spam.txt', 'r') as file:
                lines = file.readlines()
                if not lines:
                    print("Error: spam.txt is empty.")
                    return
                a, b, c = [float(value) for value in lines[0].strip().split()]
        except FileNotFoundError:
            print("Error: spam.txt not found.")
            return
        except ValueError:
            print("Error: Invalid data in spam.txt.")
            return
    else:
        print("Invalid mode selected.")
        return
    time = 1
    while True:
        try:
            temp = quadratic_solution(a, b, time, c)[1]
            print(f"Time {time}: Temperature = {temp:.2f} degrees")
        except ValueError as e:
            print(f"Error at time {time}: {e}")
            break
        choice = input("Enter 'c' to continue, 'u' to update parameters, 's' to save, or 'q' to quit: ").lower()
        if choice == 'c':
            time += 1
        elif choice == 'u':
            try:
                a = float(input("Enter new coefficient a: "))
                b = float(input("Enter new coefficient b: "))
                c = float(input("Enter new initial temperature: "))
                print("Parameters updated.")
            except ValueError:
                print("Invalid input. Parameters not updated.")
        elif choice == 's':
            with open('simulation_save.txt', 'w') as save_file:
                save_file.write(f"{a} {b} {c} {time}")
            print("Simulation state saved to 'simulation_save.txt'.")
        elif choice == 'q':
            print("Exiting simulation.")
            break
        else:
            print("Invalid choice. Continuing simulation.")
weather_model()
```

```python
#waterfall model
def quadratic_solution(a, b, time, c):
    discriminant = b**2 - 4 * a * (c - time)
    if discriminant < 0:
        raise ValueError("Negative discriminant, no real solutions.")
    sol1 = (-b - discriminant**0.5) / (2 * a)
    sol2 = (-b + discriminant**0.5) / (2 * a)
    return sol1, sol2


def requirement_definition():
    print("Stage 1: Requirement Definition")
    print("The system simulates temperatures using a quadratic equation.")
    print("Input: Coefficients a, b, c, and time range.")
    print("Output: Temperature values at specific times.\n")


def system_design():
    print("Stage 2: System Design")
    print("Design Overview:")
    print("- Use a quadratic formula to calculate temperature.")
    print("- Three modes of operation: 'hardcoded', 'input', 'file'.")
    print("- Sequential processing of time steps.\n")


def implementation(mode):
    print("Stage 3: Implementation")
    if mode == "hardcoded":
        a, b, c = 0.5, -2.0, 20.0
        return a, b, c
    elif mode == "input":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter initial temperature: "))
        return a, b, c
    elif mode == "file":
        try:
            with open('spam.txt', 'r') as file:
                lines = file.readlines()
                if not lines:
                    print("Error: spam.txt is empty.")
                    return None
                a, b, c = [float(value) for value in lines[0].strip().split()]
                return a, b, c
        except FileNotFoundError:
            print("Error: spam.txt not found.")
        except ValueError:
            print("Error: Invalid data in spam.txt.")
    return None


def verification(a, b, c):
    print("\nStage 4: Verification")
    try:
        for time in range(1, 4):
            temp = quadratic_solution(a, b, time, c)[1]
            print(f"Time {time}: Temperature = {temp:.2f} degrees")
        print("Verification successful. Outputs are consistent.")
    except ValueError as e:
        print(f"Verification failed: {e}")


def maintenance():
    print("\nStage 5: Maintenance")
    print("Options for maintenance:")
    print("- Update coefficients a, b, c.")
    print("- Rerun the simulation for new input data.")
    while True:
        choice = input("Enter 'r' to rerun, 'u' to update parameters, or 'q' to quit: ").lower()
        if choice == 'r':
            mode = input("Enter mode ('hardcoded', 'input', 'file'): ").lower()
            a, b, c = implementation(mode)
            if a is not None:
                verification(a, b, c)
        elif choice == 'u':
            try:
                a = float(input("Enter new coefficient a: "))
                b = float(input("Enter new coefficient b: "))
                c = float(input("Enter new initial temperature: "))
                verification(a, b, c)
            except ValueError:
                print("Invalid input. Parameters not updated.")
        elif choice == 'q':
            print("Exiting maintenance phase.")
            break
        else:
            print("Invalid choice. Try again.")


def waterfall_weather_model():
    requirement_definition()
    system_design()
    mode = input("Enter mode ('hardcoded', 'input', or 'file'): ").lower()
    coefficients = implementation(mode)
    if coefficients is not None:
        a, b, c = coefficients
        verification(a, b, c)
        maintenance()


waterfall_weather_model()
```
